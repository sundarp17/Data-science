hgrad and unem are independent of each other.

```
enroll_data = enroll[['unem','hgrad']].values
enroll_target = enroll[['roll']].values


X,y = scale(enroll_data), enroll_target

```

```
LinReg = LinearRegression(normalize=True)
LinReg.fit(X,y)
LinReg.score(X,y)
```
