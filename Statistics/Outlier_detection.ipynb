** Extreme value analysis using univariate methods **

```
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sb
from pylab import rcParams
```

```
'''rcparams is used for setting the size of visualization. 5,4 is length,height'''
rcParams['figure.figsize']=5,4
plt.style.use('seaborn-whitegrid')
```

```
df =pd.read_csv(r'/content/iris.data.csv')
df.columns=['sepal length','sepal width','petal length','petal width','species']
df.head()
```

```
X=df.iloc[:,0:4].values
y=df.iloc[:,4].values
```

**Identifying Outliers**

```
df.boxplot(return_type='dict')
plt.plot()
```

outliers are found at:
Sepal width > 4 and sepal width < 2.05

```
sepal_width = X[:,1]
iris_outliers = (sepal_width>4)
df[iris_outliers]
```

```
sepal_width = X[:,1]
iris_outliers = (sepal_width<2.05)
df[iris_outliers]
```

**Multivariate Outlier detection:**

```
'''we are plotting sepal length against species'''
import seaborn as sb
sb.boxplot(x='species',y='sepal length',data=df)
```
