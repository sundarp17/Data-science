** Transforming Dataset distributions **

```
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sb
from pylab import rcParams

import sklearn
from sklearn import preprocessing
from sklearn.preprocessing import scale
```

```
'''rcparams is used for setting the size of visualization. 5,4 is length,height'''
rcParams['figure.figsize']=5,4
plt.style.use('seaborn-whitegrid')
```

**Normalizing and transforming features with MinMaxScalar() and fit_transform()**

```
cars =pd.read_csv(r'/content/mtcars.csv')
cars.columns=['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
cars.head()
```

```
mpg = cars.mpg
plt.plot(mpg)
```

```
cars[['mpg']].describe()
```

```
'''make sure mpg variable in 1-column matrix.'''
'''minmax_scale will convert the data between 0 and 1'''

mpg_matrix = mpg.values.reshape(-1,1)

scaled = preprocessing.MinMaxScaler()

scaled_mpg = scaled.fit_transform(mpg_matrix) 

plt.plot(scaled_mpg)
```

```
scaled = preprocessing.MinMaxScaler(feature_range=(0,10))
scaled_mpg = scaled.fit_transform(mpg_matrix) 
plt.plot(scaled_mpg)
```

Standardizing the variables using scale function. By default, it will cente the data and scale it to unit variance.

```
stand_mpg = scale(mpg,axis=0,with_mean=False,with_std=False)
plt.plot(stand_mpg)
```

```
stand_mpg = scale(mpg)
plt.plot(stand_mpg)
```
