**Summarizing Categorical data using pandas**

There are 3 ways to describe categorical variables.
1) Counts
2) Variable Description
3) Grouping

```
import scipy
import pandas as pd
from scipy import stats
```

```
cars =pd.read_csv(r'/content/mtcars.csv')
cars.columns=['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
cars.head()
```

```
cars.index = cars.car_names
cars.head(10)
```

```
carb = cars.carb
carb.value_counts()
```

```
'''groupby'''
cars_cat = cars[['cyl','vs','am','gear','carb']]
cars_cat.head()
```

```
gears_group = cars_cat.groupby('gear')
gears_group.describe()
```

Transforming variables to categorical data type

```
'''creating a categorical variable'''
cars['group']=pd.Series(cars.gear,dtype="category")
cars['group'].dtypes
```

```
cars['group'].value_counts()
```

Describing categorical data with crosstabs

```
'''cross table shows frequency counts for features'''
pd.crosstab(cars['am'],cars['gear'])
```
