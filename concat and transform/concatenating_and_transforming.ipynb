**Concatenating and Transforming:**

Concatenating is combining data from different sources.
Transforming is converting and reformatting the data to the format that necessary for analysis.







```
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

```
df = DataFrame(np.arange(36).reshape(6,6))
df
```

```
df1 = DataFrame(np.arange(15).reshape(5,3))
df1
```

**Concatenating**

```
pd.concat([df,df1],axis=1)
```

```
pd.concat([df,df1])
```

**Transforming Data**

Dropping data

```
df.drop([0,2])
```

```
df.drop([0,2],axis=1)
```

**adding data**

```
series_obj = Series(np.arange(6))
series_obj.name = 'added_variable'
series_obj
```

```
variable_added = DataFrame.join(df, series_obj)
variable_added
```

```
added_datatable = variable_added.append(variable_added,ignore_index=False)
added_datatable
```

```
added_datatable = variable_added.append(variable_added,ignore_index=True)
added_datatable
```

**Sort Data**

```
'''sorting values in column 5'''
df_sorted = df.sort_values(by=(5),ascending=False)
df_sorted
```
